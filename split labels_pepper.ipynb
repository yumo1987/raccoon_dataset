```python
import numpy as np
import pandas as pd
np.random.seed(1)
```

```python
full_labels = pd.read_csv('outputCSV/pepper_labels.csv')
```

```python
full_labels.head(20)
```

```python
# 以filename欄位分組統計
grouped = full_labels.groupby('filename')
```

```python
# 1表示一張圖有一個label; 2表示一張圖有兩個label; 以此類推...
grouped.apply(lambda x: len(x)).value_counts()
```

### split each file into a group in a list

```python
gb = full_labels.groupby('filename')
```

```python
gb.groups
```

```python
grouped_list = [gb.get_group(x) for x in gb.groups]
```

```python
# 有標記的圖片數量
HasLabeled_Image=len(grouped_list)
print(HasLabeled_Image)
```

```python
grouped_list[0]
```

```python
# 設定分訓練/評估分割比例
val_ratio=0.2
train_image_number=int(HasLabeled_Image*(1-val_ratio))
print('train_image_number:', train_image_number)

train_index = np.random.choice(len(grouped_list), size=train_image_number, replace=False)
test_index = np.setdiff1d(list(range(HasLabeled_Image)), train_index)
```

```python
print('訓練張數:', len(train_index), ' 評估張數:', len(test_index))
```

```python
train_index[:10]
```

```python
train = pd.concat([grouped_list[i] for i in train_index])
test = pd.concat([grouped_list[i] for i in test_index])
```

```python
print('訓練標籤數量:', len(train), ' 評估標籤數量:', len(test))
```

```python
train[:5]
```

```python
train.to_csv('outputCSV/train.csv', index=None)
test.to_csv('outputCSV/test.csv', index=None)
```
